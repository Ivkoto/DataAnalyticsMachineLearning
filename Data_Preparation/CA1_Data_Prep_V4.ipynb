The function below allows us to format to format the values adding an 'M' that stands for millions or 'T' for thousands, for better visibility.

```python
def millions_formatter(x, pos):
    return f'{x / 1e6} M'

def thousands_formatter(x, pos):
    return f'{x / 1e3} K'

def standard_decimal_notation_formatter(x):
    return '{:.2f}'.format(x)
```

#### **Population Projection**

```python
filtered_bothSex = population_df[(population_df['Single Year of Age'] == 'All ages') & (population_df['Sex'] == 'Both sexes') & (population_df['Year'] >= 2015)]

#remove all the clolumns which values are just hyphens. We don't need them.
def find_all_hyphens(column):
    return not all(column == '-')
filtered_bothSex = filtered_bothSex.loc[:, filtered_bothSex.apply(find_all_hyphens)]

filtered_bothSex
```

```python
filtered_bothSex.plot(kind='scatter', x='Year', y='VALUE', figsize=(12, 6))

plt.title('Population Over the Years')
plt.xlabel('Year')
plt.ylabel('Population')

plt.xticks(rotation=90)
plt.grid(True, linestyle='--', linewidth=0.5)
plt.gca().yaxis.set_major_formatter(FuncFormatter(large_numbers_formatter))

plt.show()
```

```python
x = filtered_bothSex['Year']      
y = filtered_bothSex['VALUE']    
fit = np.polyfit(x, y, deg=1)

fit
```

```python
filtered_bothSex.plot(kind='scatter', x='Year', y='VALUE', figsize=(12, 6))

plt.title('Population')
plt.xlabel('Year')
plt.ylabel('Population')

plt.plot(x, fit[0] * x + fit[1], color='red')
plt.annotate('y={0:.0f} x + {1:.0f}'.format(fit[0], fit[1]), xy=(2000, 150000))

plt.xticks(rotation=90)
plt.grid(True, linestyle='--', linewidth=0.5)
plt.gca().yaxis.set_major_formatter(FuncFormatter(large_numbers_formatter))

plt.show()

# print out the line of best fit
'Population = {0:.0f} * Year + {1:.0f}'.format(fit[0], fit[1])
```
