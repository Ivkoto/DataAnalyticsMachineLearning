```python
from google.colab import drive
drive.mount('/content/drive/')
txt_file = "/content/drive/MyDrive/DataCleansing/university_towns.txt"


#in the steps below we convert the txt file into a DataFrame
```

```python
university_towns = []
with open('/content/drive/MyDrive/DataCleansing/university_towns.txt') as file:
  for line in file:
    if '[edit]' in line:
       # Remember this `state` until the next is found
      state = line
    else:
   # Otherwise, we have a city; keep `state` as last-seen
      university_towns.append((state, line))
university_towns[:5]
```

```python
towns_df = pd.DataFrame(university_towns, columns=['State', 'RegionName'])
towns_df.head()
```

```python
def get_citystate(item):
    if ' (' in item:
      return item[:item.find(' (')]
    elif '[' in item:
      return item[:item.find('[')]
    else:
      return item
```

```python
towns_df = towns_df.applymap(get_citystate)
```
